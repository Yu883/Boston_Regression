```python
from sklearn.datasets import load_boston
boston = load_boston()
boston
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(boston['data'], boston['target'], test_size=0.3, random_state=0)

#boston['data']>>説明変数　boston["target"]>>目的変数
```

```python
#正規化する必要がある
#X_train,X_test,y_train,y_testを使いニューラルネットの学習を行っていく
#ここはお作法的にやっていく

from sklearn.preprocessing import MinMaxScaler
scaler_x = MinMaxScaler() #インスタンス化
X_train_n = scaler_x.fit_transform(X_train)
X_test_n = scaler_x.transform(X_test)
scaler_y = MinMaxScaler()
y_train_n = scaler_y.fit_transform(y_train.reshape(len(y_train), 1))
y_test_n = scaler_y.transform(y_test.reshape(len(y_test), 1))

```

```python
#基本モデル生成
#基本モデルのクラス
from keras.models import Sequential 
model = Sequential()
#modelインスタンスを作成する　＃モデルインスタンスに中間層を加えてモデルをどんどん作成していく
```

```python
#上記modelに層を追加していく
from keras.layers import Dense #層のモジュールを読み込む
from keras.layers import core  #活性化関数のモジュールを読み込む
from keras.layers import ReLU #活性化関数発展版のモジュールを読み込む
```

```python
#中間層　modelに追加していく
model.add(
    Dense(
        128, #ユニット数
        input_shape=(13,), #入力ベクトルの次元数の指定
        activation=ReLU() #活性化関数　ReLU
    )
)
```

```python
#出力層を追加する
#bostonは回帰のため出力層は1で設定
#activation=core.activations.linearは恒等関数を読み込む
model.add(
    Dense(
        1,
        activation=core.activations.linear
    )
)
```

```python
#モデルの学習設定
from keras import losses #損失関数
from keras import optimizers #重み更新法
model.compile(
    loss=losses.mean_squared_error,   #損失関数　平均二乗誤差
    optimizer=optimizers.SGD(lr=0.01),    #重み更新法　確率的勾配降下降下法　lr=0.01学習率
)
```

```python
#モデルの学習
result = model.fit(
    X_train_n,
    y_train_n,
    epochs=100 #学習回数
)
```

```python
#予測値の算出
y_predict_n = model.predict(X_test_n)

#正規化の復元
y_predict = scaler_y.inverse_transform(y_predict_n)
print(y_predict)
```

```python
#乖離度の算出→17%乖離している
import numpy as np
np.mean(
    np.abs(y_test-y_predict.flatten())
    /y_test
)
```

```python
#基本モデル生成
from keras.models import Sequential #基本モデルのクラス
model = Sequential()
```

```python
#層の追加
from keras.layers import Dense
from keras.layers import core
from keras.layers import ReLU
```

```python
#中間層
model.add(
    Dense(
        128,
        input_shape=(13,),
        activation=ReLU()
    )
)

model.add(
    Dense(
        128,
        #input_shape=(13,), #入力ベクトルの次元数の指定
        activation=ReLU()
    )
)

model.add(
    Dense(
        128,
        #input_shape=(13,), #入力ベクトルの次元数の指定
        activation=ReLU()
    )
)

#出力層 
model.add(
    Dense(
        1,
        activation= core.activations.linear
    )
)
```

```python
#モデルの学習設定
from keras import losses #損失関数
from keras import optimizers #重み更新法
model.compile(
    loss=losses.mean_squared_error,   #損失関数　平均二乗誤差
    optimizer=optimizers.SGD(lr=0.01),    #重み更新法　確率的勾配降下降下法
)
```

```python
#モデルの学習
result = model.fit(
    X_train_n,
    y_train_n,
    epochs=100 #学習回数
)
```

```python
#予測値の算出
y_predict_n = model.predict(X_test_n)

#正規化の復元
y_predict = scaler_y.inverse_transform(y_predict_n)
print(y_predict)
```

```python
#乖離度の算出→16%乖離している
import numpy as np
np.mean(
    np.abs(y_test-y_predict.flatten())
    /y_test
)
```
